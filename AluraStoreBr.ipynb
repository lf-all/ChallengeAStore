### Importação dos dados



```python
import matplotlib.pyplot as plt
import numpy as np


import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

print(f"O Faturamento da Loja 1 é: R${faturamento_loja1}")
print(f"O Faturamento Loja 2 é: R${faturamento_loja2}")
print(f"O Faturamento Loja 3 é: R${faturamento_loja3}")
print(f"O Faturamento Loja 4 é: R${faturamento_loja4}")
```

```python
# Gráfico de barras
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
faturamentos = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]

plt.figure(figsize=(8, 6))
plt.bar(lojas, faturamentos, color=['blue', 'green', 'red', 'purple'])
plt.ylabel('Faturamento (R$)')
plt.title('Faturamento por Loja')
plt.show()

# Gráfico de pizza
faturamento_total = sum(faturamentos)
proporcoes = [(f / faturamento_total) * 100 for f in faturamentos]

plt.figure(figsize=(8, 8))
plt.pie(proporcoes, labels=lojas, autopct='%1.1f%%', startangle=140)
plt.title('Proporção do Faturamento Total por Loja')
plt.axis('equal')
plt.show()

# Gráfico de linha
plt.figure(figsize=(8, 6))
plt.plot(lojas, faturamentos, marker='o', linestyle='-')
plt.ylabel('Faturamento (R$)')
plt.title('Faturamento por Loja (Gráfico de Linha)')
plt.grid(True)
plt.show()
```

# 2. Vendas por Categoria


```python
print("\nVendas por Categoria:")
vendas_categoria_loja1 = loja['Categoria do Produto'].value_counts()
vendas_categoria_loja2 = loja2['Categoria do Produto'].value_counts()
vendas_categoria_loja3 = loja3['Categoria do Produto'].value_counts()
vendas_categoria_loja4 = loja4['Categoria do Produto'].value_counts()

print("\nLoja 1:")
print(vendas_categoria_loja1)

print("\nLoja 2:")
print(vendas_categoria_loja2)

print("\nLoja 3:")
print(vendas_categoria_loja3)

print("\nLoja 4:")
print (vendas_categoria_loja4)
```

```python
# Gráfico de barras
plt.figure(figsize=(10, 6))
vendas_categoria_loja1.plot(kind='bar', color='blue')
plt.xlabel('Categoria do Produto')
plt.ylabel('Número de Vendas')
plt.title('Vendas por Categoria do Produto na Loja 1')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Gráfico de pizza
plt.figure(figsize=(8, 8))
vendas_categoria_loja2.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Proporção das Vendas por Categoria do Produto na Loja 2')
plt.axis('equal')
plt.show()

# Gráfico de linha
plt.figure(figsize=(10, 6))
vendas_categoria_loja3.plot(kind='line', marker='o', linestyle='-')
plt.xlabel('Categoria do Produto')
plt.ylabel('Número de Vendas')
plt.title('Vendas por Categoria do Produto na Loja 3 (Gráfico de Linha)')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 3. Média de Avaliação das Lojas

```python
media_avaliacao_loja1 = loja['Avaliação da compra'].mean()
media_avaliacao_loja2 = loja2['Avaliação da compra'].mean()
media_avaliacao_loja3 = loja3['Avaliação da compra'].mean()
media_avaliacao_loja4 = loja4['Avaliação da compra'].mean()

print(f"\nMédia de Avaliação da Loja 1: {media_avaliacao_loja1:.2f}")
print(f"Média de Avaliação da Loja 2: {media_avaliacao_loja2:.2f}")
print(f"Média de Avaliação da Loja 3: {media_avaliacao_loja3:.2f}")
print(f"Média de Avaliação da Loja 4: {media_avaliacao_loja4:.2f}")
```

```python
# Gráfico de barra
avaliacoes = [media_avaliacao_loja1, media_avaliacao_loja2, media_avaliacao_loja3, media_avaliacao_loja4]

plt.figure(figsize=(8, 6))
plt.bar(lojas, avaliacoes, color=['blue', 'green', 'red', 'purple'])
plt.ylabel('Média de Avaliação')
plt.title('Média de Avaliação por Loja')
plt.ylim(0, 5)
plt.show()

# Gráfico de linha
plt.figure(figsize=(8, 6))
plt.plot(lojas, avaliacoes, marker='o', linestyle='-')
plt.ylabel('Média de Avaliação')
plt.title('Média de Avaliação por Loja (Gráfico de Linha)')
plt.ylim(0, 5)
plt.grid(True)
plt.show()

# Gráfico de dispersão
indices_lojas = np.arange(len(lojas))

plt.figure(figsize=(8, 6))
plt.scatter(indices_lojas, avaliacoes, color=['blue', 'green', 'red', 'purple'], s=100) # s é o tamanho do ponto
plt.xticks(indices_lojas, lojas) # Define os rótulos do eixo x
plt.ylabel('Média de Avaliação')
plt.title('Média de Avaliação por Loja (Gráfico de Dispersão)')
plt.ylim(0, 5)
plt.grid(True)
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python

print("\nProdutos Mais e Menos Vendidos:")

# Loja 1
print("\nLoja 1:")
vendas_produto_loja1 = loja['Produto'].value_counts()
mais_vendido_loja1 = vendas_produto_loja1.index[0]
menos_vendido_loja1 = vendas_produto_loja1.index[-1]
print(f"Produto Mais Vendido: {mais_vendido_loja1} ({vendas_produto_loja1.iloc[0]} unidades)")
print(f"Produto Menos Vendido: {menos_vendido_loja1} ({vendas_produto_loja1.iloc[-1]} unidades)")

# Loja 2
print("\nLoja 2:")
vendas_produto_loja2 = loja2['Produto'].value_counts()
mais_vendido_loja2 = vendas_produto_loja2.index[0]
menos_vendido_loja2 = vendas_produto_loja2.index[-1]
print(f"Produto Mais Vendido: {mais_vendido_loja2} ({vendas_produto_loja2.iloc[0]} unidades)")
print(f"Produto Menos Vendido: {menos_vendido_loja2} ({vendas_produto_loja2.iloc[-1]} unidades)")

# Loja 3
print("\nLoja 3:")
vendas_produto_loja3 = loja3['Produto'].value_counts()
mais_vendido_loja3 = vendas_produto_loja3.index[0]
menos_vendido_loja3 = vendas_produto_loja3.index[-1]
print(f"Produto Mais Vendido: {mais_vendido_loja3} ({vendas_produto_loja3.iloc[0]} unidades)")
print(f"Produto Menos Vendido: {menos_vendido_loja3} ({vendas_produto_loja3.iloc[-1]} unidades)")

# Loja 4
print("\nLoja 4:")
vendas_produto_loja4 = loja4['Produto'].value_counts()
mais_vendido_loja4 = vendas_produto_loja4.index[0]
menos_vendido_loja4 = vendas_produto_loja4.index[-1]
print(f"Produto Mais Vendido: {mais_vendido_loja4} ({vendas_produto_loja4.iloc[0]} unidades)")
print(f"Produto Menos Vendido: {menos_vendido_loja4} ({vendas_produto_loja4.iloc[-1]} unidades)")
```

```python

print("\nDistribuição de Preços por Loja:")

# Gráfico de histograma
plt.figure(figsize=(10, 6))
plt.hist(loja['Preço'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Preço do Produto')
plt.ylabel('Frequência')
plt.title('Distribuição de Preços dos Produtos na Loja 1')
plt.grid(axis='y', alpha=0.75)
plt.show()

# Gráfico de caixa
data_precos = [loja['Preço'], loja2['Preço'], loja3['Preço'], loja4['Preço']]

plt.figure(figsize=(10, 6))
plt.boxplot(data_precos, labels=lojas, patch_artist=True)
plt.ylabel('Preço do Produto')
plt.title('Distribuição de Preços dos Produtos por Loja')
plt.grid(axis='y', alpha=0.75)
plt.show()

# Gráfico de violino
plt.figure(figsize=(10, 6))
plt.violinplot(data_precos, showmeans=True, showmedians=True)
plt.xticks(np.arange(1, len(lojas) + 1), lojas)
plt.ylabel('Preço do Produto')
plt.title('Distribuição de Preços dos Produtos por Loja (Violin Plot)')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

# 5. Frete Médio por Loja


```python
media_frete_loja1 = loja['Frete'].mean()
media_frete_loja2 = loja2['Frete'].mean()
media_frete_loja3 = loja3['Frete'].mean()
media_frete_loja4 = loja4['Frete'].mean()

print(f"\nCusto Médio do Frete da Loja 1: R${media_frete_loja1:.2f}")
print(f"Custo Médio do Frete da Loja 2: R${media_frete_loja2:.2f}")
print(f"Custo Médio do Frete da Loja 3: R${media_frete_loja3:.2f}")
print(f"Custo Médio do Frete da Loja 4: R${media_frete_loja4:.2f}")
```

```python
# Gráfico de barras
fretes = [media_frete_loja1, media_frete_loja2, media_frete_loja3, media_frete_loja4]

plt.figure(figsize=(8, 6))
plt.bar(lojas, fretes, color=['blue', 'green', 'red', 'purple'])
plt.ylabel('Custo Médio do Frete (R$)')
plt.title('Custo Médio do Frete por Loja')
plt.show()

# Gráfico de linha
plt.figure(figsize=(8, 6))
plt.plot(lojas, fretes, marker='o', linestyle='-')
plt.ylabel('Custo Médio do Frete (R$)')
plt.title('Custo Médio do Frete por Loja (Gráfico de Linha)')
plt.grid(True)
plt.show()

# Gráfico de dispersão
plt.figure(figsize=(8, 6))
plt.scatter(indices_lojas, fretes, color=['blue', 'green', 'red', 'purple'], s=100)
plt.xticks(indices_lojas, lojas)
plt.ylabel('Custo Médio do Frete (R$)')
plt.title('Custo Médio do Frete por Loja (Gráfico de Dispersão)')
plt.grid(True)
plt.show()
```
